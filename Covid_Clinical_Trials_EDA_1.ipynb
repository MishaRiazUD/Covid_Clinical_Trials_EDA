# Conditional statements

```python
threshold = 1000

num_studies = df[df['Enrollment'] > threshold].shape[0]

print(f"The number of studies with enrollment greater than {threshold} is: {num_studies}")
```

```python
count_results = 0
for result in df['Study Results']:
    if result == 'No Results Available':
        count_results += 1
print(f"Number of clinical trials with 'No Results Available': {count_results}")
```

```python
count_results1 = 0
count_results2 = 0
for trial in df['Status']:
    if trial == 'Completed':
        count_results1 += 1
    elif trial == 'Recruiting':
        count_results2 += 1
        
print(f"Number of trials with status 'Completed': {count_results1}")
print(f"Number of trials with status 'Recruiting': {count_results2}")
```

```python
count_trials = 0
for trials in df['Conditions']:
    if trials == 'Covid-19':
        count_trials += 1
print(f"Number of clinical trials are related to 'COVID-19' : {count_trials}")
```

```python
trials_after_2020 = df[df['Start Date'] > '2020-01-01']
num_trials_after_2020 = trials_after_2020.shape[0]

print(f"The number of clinical trials that started after January 1, 2020 is: {num_trials_after_2020}")
```
